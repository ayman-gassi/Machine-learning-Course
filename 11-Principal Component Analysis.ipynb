## Les données

Travaillons de nouveau avec l’ensemble des données sur le cancer, car il avait tant de caractéristiques.

```python
from sklearn.datasets import load_breast_cancer
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
#(['DESCR', 'data', 'feature_names', 'target_names', 'target'])
```

```python
df.head()
```

## Visualisation de l’APC

Comme nous l’avons remarqué auparavant, il est difficile de visualiser des données à haute dimension, nous pouvons utiliser PCA pour trouver les deux premières composantes principales, et visualiser les données dans ce nouvel espace bidimensionnel, avec un seul diagramme de dispersion. Avant de faire cela, nous allons devoir mettre à l’échelle nos données afin que chaque fonctionnalité ait une variance unitaire unique.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
pd.DataFrame(scaled_data)
```

PCA avec Scikit Learn utilise un processus très similaire à d’autres fonctions de prétraitement qui viennent avec SciKit Learn. Nous instancions un objet PCA, trouvons les composants principaux en utilisant la méthode fit, puis appliquons la réduction de rotation et de dimensionnalité en appelant transform().

Nous pouvons également spécifier combien de composants nous voulons garder lors de la création de l’objet PCA.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

Nous pouvons maintenant transformer ces données en ses deux premières composantes principales.

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
pd.DataFrame(x_pca)
```

```python
x_pca
```

Super! Nous avons réduit 30 dimensions à seulement 2! Essayons de tracer ces deux dimensions!

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

Clearly by using these two components we can easily separate these two classes.

## Interpreting the components 

Unfortunately, with this great power of dimensionality reduction, comes the cost of being able to easily understand what these components represent.

The components correspond to combinations of the original features, the components themselves are stored as an attribute of the fitted PCA object:

Il est clair qu’en utilisant ces deux composants, nous pouvons facilement séparer ces deux classes.

## Interpréter les composants 

Malheureusement, avec ce grand pouvoir de réduction de dimensionnalité, vient le coût d’être capable de comprendre facilement ce que ces composants représentent.

Les composants correspondent à des combinaisons des caractéristiques d’origine, les composants eux-mêmes sont stockés comme un attribut de l’objet PCA ajusté :

```python
pca.components_
```

Dans ce tableau matriciel numpy, chaque ligne représente un composant principal et chaque colonne renvoie aux caractéristiques originales. nous pouvons visualiser cette relation avec une carte thermique :

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma',)
plt.show()
```

Cette heatmap et la barre de couleur représentent essentiellement la corrélation entre les différentes caractéristiques et le composant principal lui-même.
